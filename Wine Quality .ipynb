```python
import pandas as pd
import numpy as np
pd = pd.read_csv('wine-quality.csv')
pd.head(20)
```

```python
pd.isnull().sum()
```

```python
pd['quality'].unique()
```

```python
pd['quality'].describe()
```

```python
import seaborn as sns
sns.set(style="whitegrid")

ax = sns.barplot(x="quality", y="alcohol", data=pd)
```

```python
import numpy as np
conditions = [
    (pd['quality'] >= 7),
    (pd['quality'] <= 4)
]
rating = ['good', 'bad']
pd['rating'] = np.select(conditions, rating, default='average')
pd.rating.value_counts()


#Most of the wine are average...
```

```python
bx = sns.boxplot(x="quality", y='alcohol', data = pd)
bx.set(xlabel='Wine Quality', ylabel='Alcohol Percent', title='Alcohol percent in different wine quality types')


#if we increase the percentage of alcohol it would be a good wine.....
```

```python

bx = sns.boxplot(x="rating", y="pH", data = pd);
bx.set(xlabel='Wine Ratings', ylabel='pH', title='pH in different types of Wine ratings')


#If pH of wine is equale to 3.2 than we can say it is good....
```

```python
bx = sns.swarmplot(x="quality", y="pH", data = pd);
```

```python
sns.lmplot(x = "alcohol", y = "residual sugar", col = "rating", data = pd)


#we can see that in good wine residual sugar is constant when acohol is increasing....
```

```python
bx = sns.swarmplot(x="rating", y="citric acid", data = pd);

# if citric acid is greater than 0.75 we can say the wine is average...
```

```python
bx = sns.lineplot(x="quality", y="fixed acidity", data=pd)


#if fixed acidity is between 7-7.75 has more chance to be bad wine....
```

```python
bx = sns.lineplot(x="quality", y="volatile acidity", data=pd)

#if volatile acidity is between 0.34-0.38, it would be bad wine........
#if volatile acidity is between 0.26-0.30 it would be good wine.........
```
